```python
import pandas as pd
import json
import os
import numpy as np

c_id = 0
p_id = 0
claims = []
premises = []
claim_premises = []
### DebatePediaOrg Files
source_path = "C:\\Users\\Effi2\\Desktop\\debatePediaOrg\\json_files\\"
for index, file_name in enumerate(os.listdir(source_path)):
        file_path = os.path.join(source_path, file_name)
        with open(file_path, errors='ignore') as json_data:
            data = json.load(json_data)
            for claim_data in data:
                if len(claim_data['pros']) > 0 or len(claim_data['cons']) > 0:
                    claim_text = (claim_data['claimText'])
                    claim_id = c_id
                    c_id += 1
                    claims.append({"claim_id": claim_id, "claim_text": claim_text, "source": "DebatePediaOrg"})
                    for premise_data in claim_data['pros']:
                        premise_text = (premise_data['premiseText'])
                        premise_id = p_id
                        p_id += 1
                        premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "DebatePediaOrg"})
                        claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})
                    for premise_data in claim_data['cons']:
                        premise_text = (premise_data['premiseText'])
                        premise_id = p_id
                        p_id += 1
                        premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "DebatePediaOrg"})
                        claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})

claims_debatepediaorg = len(claims)       
premises_debatepediaorg = len(premises)                        

print("Length Claims DebatePediaOrg: ", claims_debatepediaorg)
print("Length Premises DebatePediaOrg: ", premises_debatepediaorg)

### DebateOrg Files
file_path = "C:\\Users\\Effi2\\Desktop\\debateOrg\\debates.json"
with open(file_path, errors='ignore') as json_data:
    data = json.load(json_data)
    for claim_data in data:
        if len(claim_data['pros']) > 0 or len(claim_data['cons']) > 0:
            claim_text = (claim_data['title'])
            claim_id = c_id
            c_id += 1
            claims.append({"claim_id": claim_id, "claim_text": claim_text, "source": "debateOrg"})
            for premise_data in claim_data['pros']:
                premise_text = (premise_data['text'])
                premise_id = p_id
                p_id += 1
                premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "debateOrg"})
                claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})
            for premise_data in claim_data['cons']:
                premise_text = (premise_data['text'])
                premise_id = p_id
                p_id += 1
                premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "debateOrg"})
                claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})                        

                
claims_debateOrg = len(claims) -     claims_debatepediaorg
premises_debateOrg = len(premises)  - premises_debatepediaorg

print("Length Claims DebateOrg: ", claims_debateOrg)
print("Length Premises DebateOrg: ", premises_debateOrg)
                
### DebateWise Files
source_path = "C:\\Users\\Effi2\\Desktop\\debatewise\\"
for index, file_name in enumerate(os.listdir(source_path)):
        file_path = os.path.join(source_path, file_name)
        with open(file_path, errors='ignore') as json_data:
            data = json.load(json_data)
            if len(data['ArgumentList']) > 0:
                claim_text = (data['MetaData']['Title'])
                claim_id = c_id
                c_id += 1
                claims.append({"claim_id": claim_id, "claim_text": claim_text, "source": "debatewise"})
                for premise_data in data['ArgumentList']:
                    argument = premise_data['Argument']
                    premise_text = (argument['Premise'][0])
                    premise_id = p_id
                    p_id += 1
                    premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "debatewise"})
                    claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})

claims_debatewise = len(claims) -     claims_debatepediaorg - claims_debateOrg
premises_debatewise = len(premises)  - premises_debatepediaorg - premises_debateOrg
print("Length Claims DebateWise: ", claims_debatewise)
print("Length Premises DebateWise: ", premises_debatewise)

### iDebateOrg Files
source_path = "C:\\Users\\Effi2\\Desktop\\iDebateOrg\\json_files"
for index, file_name in enumerate(os.listdir(source_path)):
        file_path = os.path.join(source_path, file_name)
        with open(file_path, errors='ignore') as json_data:
            data = json.load(json_data)
            if len(data['pros']) > 0 or len(data['cons']) > 0:
                claim_text = (data['title'])
                claim_id = c_id
                c_id += 1
                claims.append({"claim_id": claim_id, "claim_text": claim_text, "source": "iDebateOrg"})
                for premise_data in data['pros']:
                    premise_text = (premise_data['title point pro claim'])
                    premise_id = p_id
                    p_id += 1
                    premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "iDebateOrg"})
                    claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})
                for premise_data in data['cons']:
                    premise_text = (premise_data['text point con claim'])
                    premise_id = p_id
                    p_id += 1
                    premises.append({"premise_id": premise_id, "premise_text": premise_text, "source": "iDebateOrg"})
                    claim_premises.append({"claim_id": claim_id, "premise_id": premise_id})                        



claims_DebateOrg = len(claims) -     claims_debatepediaorg - claims_debateOrg - claims_debatewise
premises_DebateOrg = len(premises)  - premises_debatepediaorg - premises_debateOrg - premises_debatewise

print("Length Claims DebateOrg: ", claims_DebateOrg)
print("Length Premises DebateOrg: ", premises_DebateOrg)                        

premises_df = pd.DataFrame(premises)
claims_df = pd.DataFrame(claims)
claim_premises_df = pd.DataFrame(claim_premises)
```
